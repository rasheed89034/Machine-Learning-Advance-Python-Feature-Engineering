## Handling Outliers Using Python
### Introduction to Box Plots and Five-Number Summary
##### Q1, how to calculate Q1 (which is the 25th percentile), and the five-number summary, which includes the minimum, maximum, Q1, Q3, and median values. These five values are essential for creating a box plot.

```python
import numpy as np 
```

```python
marks = [45, 32, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74]
minimum,Q1,median,Q3,maximum = np.quantile(marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR = (Q3-Q1)
print(IQR)
```

```python
## This is way to find the outliers manually 
lowerFence = Q1 - 1.5 *(IQR)
higherFence = Q3 + 1.5*(IQR)
```

```python
print(lowerFence)
print(higherFence)
```

```python
import seaborn as sns
```

```python
# this is a way to find the outliers using seaborn library
sns.boxplot(marks)
```

```python
# now adding some outliers 
marks_outliers = marks + [-100,-150,-200]
sns.boxplot(marks_outliers)
```
